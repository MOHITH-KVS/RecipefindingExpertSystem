function to match the recipes

```python
def find_recipe(user_ingredients,recipes):
  matching_recipes = []
  #checking all the recipes in the recipes list by using "recipe" variable and assuming a variable "all_user_ingredients"
  #be true that all the ingredients with user and my ingredients data is same
  for recipe in recipes:
    all_user_ingredients = True
    #Now checking the ingredients with user_ingredients to check whether our assumption is write or wrong
    for ingredient in recipe['ingredients']:
      if ingredient not in user_ingredients:
        all_user_ingredients = False
        break
        #if the all_user_ingredients are true then this if statement will execute and displays the recipe name and preparation time (approxS)
    if all_user_ingredients:
           matching_recipes.append({
             'recipe_name': recipe['recipe_name'],
             'prep_time': recipe.get('prep_time','N/A')
         })
      #finally returning the matched recipe name
  return matching_recipes
```

function to get their leftover food ingredients from the user

```python
#taking the user_having_function to get the available ingredient from the user
def user_having_ingredients():
  #the below statements in the printf will be displayed
  print("Hello Foodie!")
  print("Enter your available food ingredients to known the recipes (comma-separated):")
  #the split function is used to separated with commas
  user_input = input()
  return [ingredient.strip() for ingredient in user_input.split(',')]
```

```python
user_ingredients = user_having_ingredients()

# Find and display matching recipes
suggested_recipes = find_recipe(user_ingredients, recipes)
if suggested_recipes:
    print("You can make the following recipes with your ingredients:")
    for recipe in suggested_recipes:
        print(f"- {recipe['recipe_name']} (\nApprox. time: {recipe['prep_time']}minutes)")
else:
    print("No recipes found with the given ingredients.")
```
